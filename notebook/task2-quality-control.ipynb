```python
import pandas as pd
import json
import csv
import ast
```

```python
t2_df = pd.read_csv("../data/t2_guess/output/task2-guess-final.csv")
```

```python
worker_dictionary = {}
cols = ["000000022892", "000000402433", "000000426166", "000000576955", "000000577862"]
answer_cols = ["cat", "pizza", "bicycle", "horse", "giraffe"]
for index, row in t2_df.iterrows():
    right = 0
    total = 0
    worker_id = row["WorkerId"]
    for i in range(len(cols)):
        guesses = ast.literal_eval(row[str(cols[i])])
        if answer_cols[i] in guesses:
            right += 1
            total += 1
        else:
            total += 1
    if worker_id in worker_dictionary:
        old_right = worker_dictionary.get(worker_id)
        new_right = [old_right[0] + right, old_right[1] + total]
        worker_dictionary[worker_id] = new_right
    else:
        worker_dictionary[worker_id] = [right, total]

lst = [(x, round(i[0]/i[1], 3)) for x, i in worker_dictionary.items()]
task2_csv = sorted(lst, key = lambda x: (x[1]), reverse=True)

with open("../quality/task2-quality.csv", "w") as o:
    csv_out = csv.writer(o)
    csv_out.writerow(["worker_id", "quality"])
    csv_out.writerows(task2_csv)
        
```
